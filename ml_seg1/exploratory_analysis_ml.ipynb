```python
# import dependencies
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
# read in csv as dataframe
df = pd.read_csv("county_statistics_for_model.csv", index_col=0)
df.head()
```

```python
# drop unnecessary rows
df = df.drop(['percentage16_Donald_Trump', 'percentage16_Hillary_Clinton', 'total_votes16', 'votes16_Donald_Trump', 'votes16_Hillary_Clinton', 'lat', 'long'], axis = 1)
df.columns
```

```python
# drop rows with null values and check if there are any nulls in columns after
df = df.dropna(axis = 0, how = 'any')
df.isnull().sum()
```
